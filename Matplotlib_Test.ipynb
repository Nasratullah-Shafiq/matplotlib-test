###Bar pLot
Bar plot show the categorical data.

```python
categories = ['HR', 'Finance', 'System Development', 'Purchase', 'Sale', 'Accounting']
values = [23, 34, 65, 65, 55, 66]

plt.bar(categories, values, color = 'green')
plt.title('Employee by Departments')
plt.xlabel('Departments')
plt.ylabel('Number of Employees')
plt.show()
```

###Scatter Plot
scatter plot define the relationship between variables

```python
x = [5, 7, 4, 8, 9]
y = [12, 15, 17, 45, 45]

plt.scatter(x, y, color = 'red')
plt.show()
```
